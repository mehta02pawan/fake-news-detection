```python
from sklearn.model_selection import train_test_split
# Split the dataset
labels = df.label
x_train,x_test,y_train,y_test=train_test_split(df['text'], labels, test_size=0.2, random_state=7)
print(x_train.head)
```

```python
#Initialize a TfidfVectorizer

from sklearn.feature_extraction.text import TfidfVectorizer,TfidfTransformer

tfidf_vectorizer=TfidfVectorizer(stop_words='english', max_df=0.7)
#DataFlair - Fit and transform train set, transform test set
tfidf_xtrain=tfidf_vectorizer.fit_transform(x_train)
# tfidf_xtrain_Gaussian=tfidf_vectorizer.fit_transform(x_train.to_xarray())
# tfidf_ytrain=tfidf_vectorizer.fit_transform(y_train) 
tfidf_test=tfidf_vectorizer.transform(x_test)

print(tfidf_xtrain.shape)
print(tfidf_test.shape)
print(tfidf_test)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split, GridSearchCV, RandomizedSearchCV
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import LogisticRegression
```

```python
model_params = {
    # 'GaussianNB':{
    #      'model':GaussianNB()
    #  },
    'DecisionTree':{
        'model':DecisionTreeClassifier(criterion='gini'),'params':{'max_depth':[7,9,11] }
    },
    'RandomForest':{
        'model':RandomForestClassifier(),'params':{'n_estimators':[20,50,70]}
    },
    'KNN':{
        'model': KNeighborsClassifier(),'params':{'n_neighbors':[10,20,30]}
    },
    'LogisticRegression':{
        'model':LogisticRegression(),'params':{'max_iter':[50,100,150,200,250,300,350]}
    }
}
```

```python
#In this loop, i have trained model for different classfiers for different number of parameters and store accuracy of all in scores list.
scores = []
for model_name,mp in model_params.items():
    clf = RandomizedSearchCV(mp['model'],mp['params'],cv=2,return_train_score=False,n_iter=2)
    clf.fit(tfidf_xtrain,y_train)
    scores.append({
        'model':model_name,
        'best_score':clf.best_score_,
        'best_param':clf.best_params_
    })
```

```python
print(scores)
```

```python
#This is the best accuracy from all classifiers listed above.
final_df = pd.DataFrame(scores,columns=['model','best_score','best_param'])
print(final_df)
```
